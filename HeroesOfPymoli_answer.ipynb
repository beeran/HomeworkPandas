```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
#Display the total number of players
player_count = purchase_data["SN"].unique()
player_df = pd.DataFrame({"Total Players":[len(player_count)]})
player_df.head()
```

```python
unique_item = purchase_data["Item ID"].unique()
avg_price = purchase_data["Price"].mean()
count_purchases = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum();
raw_data = {"Number of Unique Items":[len(unique_item)], 
            "Average Price":[f'${avg_price:.2f}'], 
            "Number of Purchases":[count_purchases], 
            "Total Revenue":[f'${total_revenue:,}']}
item_summary = pd.DataFrame(raw_data) 
item_summary.head()
```

```python
def print_pct(x): return f'{100 * x:.2f}'
gender_grouped = purchase_data.groupby("Gender")
gender_unique_count = gender_grouped["SN"].nunique().rename("Total Count")
total_unique_count = gender_unique_count.sum()
gender_pct = (gender_unique_count / total_unique_count).apply(print_pct).rename("Percent of Players")
result = pd.concat([gender_unique_count, gender_pct], axis = 1).sort_values("Total Count", ascending = False)
result
```

```python
print_currency = lambda x : f'${x:,.2f}'
def format_currency_column(data_frame, column_names):
    for column_name in column_names:
        data_frame[column_name] = data_frame[column_name].apply(print_currency)
    return

gender_grouped = purchase_data.groupby("Gender")
purchase_count = gender_grouped["Purchase ID"].count().rename("Purchase Count")
avg_price = gender_grouped["Price"].mean().apply(print_currency).rename("Average Purchase Price")
total_price = gender_grouped["Price"].sum().rename("Total Purchase Value")
unique_sn = gender_grouped["SN"].nunique();
avg_price_per_person = (total_price / unique_sn).rename("Avg Total Purchase per Person")
result = pd.concat([purchase_count, avg_price, total_price, avg_price_per_person], axis = 1)
format_currency_column(result, ["Total Purchase Value", "Avg Total Purchase per Person"])
result
```

```python
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_binned = purchase_data
age_binned["Age Range"] = pd.cut(age_binned["Age"], age_bins, labels = age_labels)
age_bin_grouped = age_binned.groupby("Age Range")
unique_sn = age_bin_grouped["SN"].nunique().rename("Total Count");
pct = (unique_sn / total_unique_count).apply(print_pct).rename("Percentage of Players")
result = pd.concat([unique_sn, pct], axis = 1)
result

```

```python
purchase_count = age_bin_grouped["SN"].count().rename("Purchase Count");
avg_price = age_bin_grouped["Price"].mean().rename("Average Purchase Price");
total_price = age_bin_grouped["Price"].sum().rename("Total Purchase Value");
unique_sn = age_bin_grouped["SN"].nunique();
avg_price_per_person = (total_price / unique_sn).rename("Avg Total Purchsae per Person");

result = pd.concat([purchase_count, avg_price, total_price, avg_price_per_person], axis = 1)
format_currency_column(result, ["Average Purchase Price", "Total Purchase Value", "Avg Total Purchsae per Person"])
result
```

```python
sn_grouped = purchase_data.groupby("SN")
purchase_count = sn_grouped["Purchase ID"].count().rename("Purchase Count")
avg_price = sn_grouped["Price"].mean().rename("Average Purchase Price")
total_price = sn_grouped["Price"].sum().rename("Total Purchase Value")
result = pd.concat([purchase_count, avg_price, total_price], axis = 1).sort_values("Total Purchase Value", ascending = False)
format_currency_column(result, ["Average Purchase Price", "Total Purchase Value"])
result.head()


```

```python
item_grouped = purchase_data.groupby("Item ID")
item_name = item_grouped["Item Name"].first()
purchase_count = item_grouped["Purchase ID"].count().rename("Purchase Count")
item_price = item_grouped["Price"].first().rename("Item Price")
total_price = item_grouped["Price"].sum().rename("Total Purchase Value")
item_result = pd.concat([item_name, purchase_count, item_price, total_price], axis = 1)
result = item_result.sort_values("Purchase Count", ascending = False)
format_currency_column(result, ["Item Price", "Total Purchase Value"])
result.head()
```
